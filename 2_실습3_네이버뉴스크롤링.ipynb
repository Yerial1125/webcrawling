### 2. 네이버 헤드라인 뉴스 링크 저장하기

- 헤드라인 뉴스들의 제목, 링크, 내용을 출력하고, 데이터프레임으로 저장하기
- 저장된 데이터프레임을 csv 파일로 저장하기

```python
url = 'https://news.naver.com'
html = urlopen(url)
```

```python
soup = BeautifulSoup(html, 'html.parser')
soup
```

```python
link = soup.select('.cjs_news_a')

link_list = []
for i in link :
    link_list.append(i['href'])
    
link_list
```

```python
title_list = []
content_list = []

def news_func(url) :
    url = url
    html = urlopen(url)
    soup = BeautifulSoup(html, 'html.parser')
    title = soup.select('#title_area span')
    content = soup.select('._article_content') 
        
    for i in title :
        title_list.append(i.text)
    for i in content :
        content_list.append(i.text)
```

```python
for link in link_list:
    news_func(link)
```

```python
import pandas as pd

naver_news = pd.DataFrame([title_list, content_list, link_list]).T
naver_news.rename(columns={0:'제목',1:'내용',2:'링크'}, inplace=True)
naver_news
```

```python
naver_news.to_csv('naver_news.csv')
```
